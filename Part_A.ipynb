```python
import nltk
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import TfidfVectorizer

nltk.download('punkt')
```

```python
document = """
Artificial Intelligence is transforming industries.
Machine learning enables computers to learn from data.
Deep learning is a subset of machine learning.
Neural networks are used in deep learning.
AI is used in healthcare, finance, and education.
Large language models are becoming popular.
"""
```

```python
def sliding_window(text, window_size=200, overlap=50):
    chunks = []
    start = 0

    while start < len(text):
        end = start + window_size
        chunks.append(text[start:end])
        start += window_size - overlap

    return chunks

chunks = sliding_window(document)

print("Chunks:")
for i, chunk in enumerate(chunks):
    print(f"\nChunk {i+1}:\n{chunk}")
```

```python
raw_text_layer = chunks
```

```python
def summarize(chunk):
    sentences = chunk.split(".")
    return sentences[0]

summary_layer = [summarize(chunk) for chunk in raw_text_layer]

print(summary_layer)
```

```python
def categorize(text):
    if "AI" in text or "Artificial Intelligence" in text:
        return "Artificial Intelligence"
    elif "machine learning" in text:
        return "Machine Learning"
    elif "deep learning" in text:
        return "Deep Learning"
    else:
        return "General"

category_layer = [categorize(text) for text in summary_layer]

print(category_layer)
```

```python
def extract_keywords(text):
    words = text.split()
    return words[:5]

distilled_layer = [extract_keywords(text) for text in summary_layer]

print(distilled_layer)
```

```python
knowledge_pyramid = []

for i in range(len(raw_text_layer)):
    knowledge_pyramid.append({
        "raw": raw_text_layer[i],
        "summary": summary_layer[i],
        "category": category_layer[i],
        "distilled": distilled_layer[i]
    })

print(knowledge_pyramid)
```

```python
def retrieve(query):
    corpus = summary_layer + raw_text_layer
    vectorizer = TfidfVectorizer()

    vectors = vectorizer.fit_transform(corpus + [query])

    similarity = cosine_similarity(
        vectors[-1],
        vectors[:-1]
    )

    index = similarity.argmax()

    return corpus[index]

query = "What is machine learning?"
result = retrieve(query)

print("Query:", query)
print("Result:", result)
```

```python
while True:
    query = input("Ask question: ")
    print(retrieve(query))
```
